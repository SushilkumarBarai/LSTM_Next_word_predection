```python
faqs = """Certainly, here's a detailed description of the iPhone 13 Pro in paragraph form:

**Design:** The iPhone 13 Pro showcases a premium design that exudes elegance and durability. Crafted with a stainless steel frame, it combines aesthetics with strength. The front cover features Apple's ceramic shield technology for added protection. You have a choice of sophisticated finishes, including Graphite, Gold, Silver, and the striking Sierra Blue.

**Display:** At the heart of the iPhone 13 Pro is its Super Retina XDR display, which employs ProMotion technology to deliver both high resolution and a high refresh rate. This results in a visually stunning screen, typically around 6.1 inches in size, offering smooth scrolling, fluid animations, and responsive touch interactions.

**Performance:** Powering the iPhone 13 Pro is Apple's formidable A15 Bionic chip, renowned for its impressive performance and energy efficiency. Whether you're engaging in intensive gaming, multitasking, or resource-demanding applications, this smartphone is designed to handle it all with finesse.

**Camera System:** The iPhone 13 Pro boasts a sophisticated camera system, featuring a triple-lens setup on the rear. This setup typically includes an ultra-wide, wide, and telephoto lens. These lenses enable advanced photography and videography, with features like ProRAW for photography enthusiasts and ProRes video recording for professional-grade video content creation.

**Battery Life:** Apple continually enhances battery life in its iPhone series. The iPhone 13 Pro is no exception, aiming to provide all-day battery life under normal usage conditions. This ensures that you can stay connected and productive throughout your day without constantly seeking a charger.

**Software:** Running on iOS, Apple's proprietary mobile operating system, the iPhone 13 Pro benefits from regular updates and access to a vast ecosystem of applications and services. The intuitive user interface and seamless integration with other Apple devices contribute to a cohesive and user-friendly experience.

**Connectivity:** Embracing the era of 5G, the iPhone 13 Pro offers faster data speeds and broader connectivity options. It supports Wi-Fi 6 for efficient wireless networking, Bluetooth for seamless pairing with other devices, and NFC for contactless payments and interactions.

**Storage:** The iPhone 13 Pro is available in various storage configurations, ranging from 128GB to a capacious 1TB. This diversity in storage options ensures that you have ample space for your apps, photos, videos, and other content.

**Additional Features:** The iPhone 13 Pro incorporates a suite of additional features, including Face ID for secure facial recognition-based authentication, water and dust resistance with an IP68 rating, and compatibility with a range of accessories like the Apple Pencil and MagSafe, which enables wireless charging and attachment of various accessories.

It's important to note that the specific specifications and features of the iPhone 13 Pro may vary based on the model and the region where it is sold. To obtain the most up-to-date and precise information, it's recommended to visit the official Apple website or consult an authorized Apple retailer."""
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
len(tokenizer.word_index)
```

```python
word_index_class=len(tokenizer.word_index)
word_index_class
```

```python
input_sequences = []
for sentence in faqs.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
max_len
```

```python

padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python

y = to_categorical(y,num_classes=278)    #tokeizer output start with 1 not with zero thats why i ma adding +1
```

```python
y.shape
```

```python


model = Sequential()

model.add(Embedding(input_dim=278, output_dim=100, input_length=53))

model.add(LSTM(150, return_sequences=True))

model.add(LSTM(150))

model.add(Dense(278, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
model.fit(X,y,epochs=100)
```

```python

text = "The iPhone 13 Pro showcases a"
token_text = tokenizer.texts_to_sequences([text])[0]
padded_token_text = pad_sequences([token_text], maxlen=53, padding='pre')
  # predict
pos = np.argmax(model.predict(padded_token_text))
for word,index in tokenizer.word_index.items():
  if index == pos:
    print(word)
```

```python

text = "what is the iphone"

for i in range(10):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=53, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
